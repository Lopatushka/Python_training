```python
# Open file
genome = open('sequence.fasta')
```

```python
# Read line
s1 = genome.readline()
s1

# Close file
genome.close()
```

```python
# Open file and close it after operations
# Strip() - delete \n \t etc...

with open('sequence.fasta') as genome:
    s1 = genome.readline().strip()
s1
    
```

```python
# Print all lines from the file

with open('sequence.fasta') as genome:
    for i in genome:
        i = i.strip()
        print(i)
```
